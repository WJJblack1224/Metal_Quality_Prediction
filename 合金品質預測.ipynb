#載入模型進行訓練

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.2,random_state=17,shuffle=True,stratify=y)
```

```python
clf = DecisionTreeClassifier(max_depth=3)
clf.fit(train_X, train_y)
predictions = clf.predict(test_X)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid = {'n_estimators': [50, 100, 200], 'max_features': ['auto', 'sqrt'], 'bootstrap': [True, False], 'criterion':['entropy', 'gini']}
rfcgrid = GridSearchCV(RandomForestClassifier(random_state=101), param_grid, verbose=100, cv=10, n_jobs=-2)
rfcgrid.fit(train_X, train_y)
```

```python
# check the best params
rfcgrid.best_params_
```

```python
from sklearn.ensemble import GradientBoostingClassifier

param_grid = {'n_estimators':[100, 200, 300], 'loss' : ['log_loss', 'exponential'], 'learning_rate':[0.001, 0.01, 0.1, 1, 10], 'criterion':['friedman_mse', 'mse']}
grid = GridSearchCV(GradientBoostingClassifier(), param_grid, verbose=True, cv=10, n_jobs=-2)
grid.fit(train_X, train_y)
```

```python
# check the best params
grid.best_params_
```

```python
from xgboost import XGBClassifier

param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.001, 0.01, 0.1, 1, 10],
    'max_depth': [3, 4, 5],
    'gamma': [0, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
}

xgb_model = XGBClassifier()
xgbgrid = GridSearchCV(xgb_model, param_grid, verbose=True, cv=10, n_jobs=-2)
xgbgrid.fit(train_X, train_y)
```

```python
# check the best params
xgbgrid.best_params_
```

#神經網路模型(DNN)

```python
import numpy as np
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

資料預處理

```python
print(train_y.shape)
```

```python
train_y = train_y.values.reshape(-1, 1)
test_y = test_y.values.reshape(-1, 1)
```

```python
print(train_X.shape,train_y.shape)
```

```python
from keras.utils import to_categorical

# 將標籤進行 one-hot 編碼
train_y = to_categorical(train_y, num_classes=5)
test_y = to_categorical(test_y, num_classes=5)
```

建立神經網路

```python
#建立模型
model = Sequential([
    Dense(300,input_dim=27,activation='relu'),
    Dropout(0.3),
    Dense(50,activation='relu'),
    Dropout(0.3),
    Dense(5,activation='softmax')
])
```

```python
# 編譯模型
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping

# 建立早停回調函數
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
# 訓練模型並保存訓練過程中的損失函數值
history = model.fit(train_X, train_y, epochs=100, batch_size=50, validation_split=0.1, verbose=1,callbacks=[early_stopping])

# 獲取訓練過程中的損失函數值和驗證損失函數值
losses = history.history['loss']
val_losses = history.history['val_loss']
epochs = range(1, len(losses) + 1)

# 繪製訓練損失函數值和驗證損失函數值的變化
plt.plot(epochs, losses, label='Training Loss')
plt.plot(epochs, val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

```python
DNN_predictions = model.predict(test_X)
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np

# 將 one-hot 編碼的標籤轉換為類別標籤
test_y_labels = np.argmax(test_y, axis=1)
DNN_predictions_labels = np.argmax(DNN_predictions, axis=1)

# 計算混淆矩陣
cm = confusion_matrix(test_y_labels, DNN_predictions_labels)

class_names = ["grade0", "grade1", "grade2", "grade3", "grade4"]
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Reds", xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predictions')
plt.ylabel('True Values')
plt.title('Confusion Matrix')
plt.show()
```
